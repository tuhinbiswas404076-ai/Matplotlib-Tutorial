# Scatter Plot:
A Scatter Plot shows the relationship between two numerical variables.
Each point (dot) represents one observation in the dataset.

1. Used in Bivariate Analysis

# groupby-(Pandas tool)-

1. Summarize data by category
2. Compute statistics (mean, sum, count, etc.) per group
3. Split-Apply-Combine strategy:
4. Split → divide data into groups
5. Apply → apply a function (mean, sum, count)
6. Combine → results returned in a new DataFrame


```python
# Calcutale department salary

salary_by_dept = df.groupby("dept")["Salary"].sum()
salary_by_dept
```

```python
# pie chat
plt.pie(salary_by_dept, labels = salary_by_dept.index, autopct= "%1.2f")
plt.axis("equal")
plt.show()
```

shadow = True
1. Adds a shadow effect below the pie chart
2. Makes the chart look 3D / visually appealing

```python
plt.pie(salary_by_dept, labels = salary_by_dept.index, colors = ["red", "pink", "black"], autopct= "%1.1f",shadow = True, explode= [0, 0.1, 0.2])
plt.axis("equal")
plt.show()
```

# Bubble Plot:
A Bubble Plot is an extension of a Scatter Plot used in Multivariate Analysis.

## It shows three variables at the same time:
1. X-axis → first variable
2. Y-axis → second variable
3. Bubble size → third variable
4. Sometimes color represents a fourth variable.

```python
# Here 3 numerical column present.

plt.scatter(df["Age"], df["Salary"], s = df["experience"], color = "red")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age VS Salary Vs experience")
plt.show()

# by defalt here bubble ar too small...Let's solve the problem
```

```python
plt.scatter(df["Age"], df["Salary"], s = df["experience"]*25, color = "red", edgecolors= "black")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age VS Salary Vs experience")
plt.show()

# (s = df["experience"]*25) -> multipling by 25 we improve size
# (edgecolors= "black") -> it get black color of edge of bubble
```

# Bubble plot:
for (2 numerical and 1 catagorical column)

```python
plt.scatter(df['Age'], df['Salary'], c = df['dept'].map({"HR": "red", "IT": "blue", "Finance": "yellow"}), edgecolors= "black")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age VS Salary Vs Dept")
plt.legend()
plt.show()

# But here in ploat can not mention which color represent which data.
# Let's solve it.
```

```python
# Mention color first
color = {"HR": "red", "IT": "blue", "Finance": "yellow"}

for dept, color in color.items():
  df_dept = df[df["dept"] == dept]
  plt.scatter(df_dept["Age"], df_dept["Salary"], c = color, edgecolors= "black", label = dept)

plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age VS Salary Vs Dept")
plt.legend()
plt.show()
```

# Object-Oriented API-
1. Object-Oriented API is a plotting method in Matplotlib used to create and control graphs using Figure and Axes objects.
2. Figure (fig) represents the entire plotting window.
3. Axes (ax) represents the area where the graph is drawn.
4. The basic syntax is:            fig, ax = plt.subplots()
5. In this method we use functions like ax.plot(), ax.scatter(), ax.bar() instead of plt.plot().
6. It provides better control over plots and is useful for multiple graphs in one figure.
7. It is commonly used for complex and professional data visualizations.

```python
fig, axs = plt.subplots(1,3, figsize = (10, 5))

# plt.subplots()-> used to create a figure and multiple subplots (axes).
# fig → Figure object, which is the entire plotting window.
# axis → represents the Axes objects, where the graphs will be drawn.
# (1,3 )→ means 1 row and 3 columns, So this code creates 3 graphs in one row.
# figsize=(10,5) → sets the size of the figure: Width = 10, Height = 5
# axis becomes a list-like array of 3 axes:
```

```python
from matplotlib import markers
# Let full the above three empy plot using line, histogram, Boxplot
fig, axs = plt.subplots(1, 3, figsize = (20, 5))

# Line plot:

axs[0].plot(sort_age["Age"], df["Salary"], color = "red", marker = "*", linewidth = "2", markersize = 2)
axs[0].grid()
axs[0].set_title("Line Plot")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Salary")

# Histogram:

axs[1].hist(df['Salary'], bins = 5, color = "yellow")
axs[1].set_title("Histogram")
axs[1].set_xlabel("Salary")
axs[1].set_ylabel("Frequency")

# Box plot:

axs[2].boxplot(df["Salary"])
axs[2].set_title("Box Plot")
axs[2].set_xlabel("Salary")
axs[2].set_ylabel("Salary")

plt.savefig("plot.png") # It save the image in laptop inviroment
plt.show()
```
